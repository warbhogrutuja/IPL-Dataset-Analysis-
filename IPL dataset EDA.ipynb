```python
#Step 1: Load All Tables
import pandas as pd
import os

folder = r"C:\Users\Rutuja\.cache\kagglehub\datasets\krishd123\ipl-2026-complete-dataset\versions\13"

matches = pd.read_csv(f"{folder}\\matches.csv")
deliveries = pd.read_csv(f"{folder}\\deliveries.csv")
batting = pd.read_csv(f"{folder}\\batting_stats.csv")
bowling = pd.read_csv(f"{folder}\\bowling_stats.csv")
fielding = pd.read_csv(f"{folder}\\fielding_stats.csv")
points = pd.read_csv(f"{folder}\\points_table.csv")
squads = pd.read_csv(f"{folder}\\squads.csv")
venues = pd.read_csv(f"{folder}\\venues.csv")
```

```python
#Step 2: Data Quality Check
datasets = {
    "matches": matches,
    "deliveries": deliveries,
    "batting": batting,
    "bowling": bowling,
    "fielding": fielding,
    "points": points,
    "squads": squads,
    "venues": venues
}

for name, df in datasets.items():
    print("\n" + "="*50)
    print(name.upper())

    print("Shape:", df.shape)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicates:")
    print(df.duplicated().sum())

    print("\nData Types:")
    print(df.dtypes)
```
